```python
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn


def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
database = r"./wiki.db"

sql_create_wiki_table = """ CREATE TABLE IF NOT EXISTS wiki (
                                    id integer PRIMARY KEY,
                                    sentence text NOT NULL,
                                    target integer NOT NULL,
                                    parent integer, 
                                    sequence integer NOT NULL
                                ); """

sql_create_wiki_augmented_table = """CREATE TABLE IF NOT EXISTS wiki_augmented (
                                    id integer PRIMARY KEY,
                                    augmented_sentence text NOT NULL,
                                    target integer NOT NULL,
                                    wiki_id integer NOT NULL,
                                    sequence integer NOT NULL,
                                    FOREIGN KEY (wiki_id) REFERENCES wiki (id)
                                );"""

# create a database connection
conn = create_connection(database)

# create tables
if conn is not None:
    # create projects table
    create_table(conn, sql_create_wiki_table)

    # create tasks table
    create_table(conn, sql_create_wiki_augmented_table)
else:
    print("Error! cannot create the database connection.")
```
